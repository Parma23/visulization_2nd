### Task 2: Load the data
Write a function which does the following:
- Input: 
    - Function should take a filepath (str) object
- Creates a dataframe using pandas
- Lower-case all column headers
- Replace the space with an underscore (_) in all column headers
- Output:
    - Cleaned dataframe object

```python
def load_data(file):
    data = pd.read_csv(file)
    for column in data.columns:
        column.lower()
        column.replace(" ", "_")
    return data


df = load_data(r'C:\Users\xende\Downloads\Food_Costs.csv')
```

- Create a visualization detailing the ratio of non-null values to null values in the entire dataset

```python
null_counts = df.isnull().sum()
non_null_counts = df.notnull().sum()
ratios = non_null_counts / null_counts
fig = px.bar(x=ratios.index, y=ratios.values, labels={'x': 'Columns', 'y': 'Non-null Ratio'},
             title='Ratio of Non-null Values to Null Values')
fig.update_traces(marker_color='lightblue', marker_line_color='black', marker_line_width=1)
fig.update_layout(
    xaxis_title_font=dict(family='Arial', size=12),
    yaxis_title_font=dict(family='Arial', size=12),
    font=dict(family='Arial', size=10)
)
```

#### *According to the data visually represented above, we see that the number of Non-Null Values is directly correlated to Null Values.*
